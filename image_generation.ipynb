```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A chai in a traditional cup on a wet surface and having a rose flower boquet placed left side of the cup on the same surface, HD"
image = pipe(prompt).images[0]
image
```

```python
image = pipe(
    prompt,
    negative_prompt="blurry, low quality"
).images[0]
image
```

```python
image = pipe(
    prompt,
    height=512,
    width=768
).images[0]
image
```

```python
prompt = "The nature on rainy day, HD"

#Normal Image
image1 = pipe(prompt).images[0]
display(image1)

#Negative Prompt Image
image2 = pipe(prompt, negative_prompt="blurry, low quality").images[0]
display(image2)

#Different Aspect Ratio
image3 = pipe(prompt, height=512, width=768).images[0]
display(image3)
```
